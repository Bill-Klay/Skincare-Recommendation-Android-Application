### Tokeninzing and cleaning the reviews

```python
from nltk import word_tokenize
from nltk.corpus import stopwords
import pandas as pd
import string
import pyodbc
import joblib
```

```python
df = pd.read_excel('reviews.xlsx')
stop = set(stopwords.words('english') + list(string.punctuation))
df
```

```python
tokens = []
for i in df['Review_cleaned']:
    tokens.append([j for j in word_tokenize(i.lower()) if i not in stop])
print(tokens)
```

### Firbase connection and data upload

```python
import firebase_admin
from firebase_admin import firestore
from firebase_admin import credentials
import pandas as pd
```

```python
cred = credentials.Certificate("serviceAccountKey.json")
firebase_admin.initialize_app(cred)
db = firestore.client()
```

```python
productDF = pd.read_excel("product_data_sliced.xlsx")
productDF
```

```python
ing = productDF.ingredients[productDF['product_id'] == 100].item()
ing
```

```python
for i in range(0, len(productDF)):
    ing = productDF.ingredients[productDF['product_id'] == productDF['product_id'][i]].item()
    print(productDF['product_id'][i])
    print(ing)
```

```python
for i in range(0, len(productDF)):
    ing = productDF.ingredients[productDF['product_id'] == productDF['product_id'][i]].item()
    data = {'ingredients': ing}
    try:
        db.collection('products').document(str(productDF['product_id'][i])).update(data)
    except:
        print('No document available for ID {}'.format(productDF['product_id'][i]))
```

```python
data = {}
for i in range(0, len(productDF)):
    data = {
        'product_id': int(productDF['product_id'][i]),
        'product_name': str(productDF['product_name'][i]),
        'product_url': str(productDF['product_url'][i]),
        'image_url': str(productDF['image_url'][i]),
        'category': str(productDF['category'][i]),
        'price': float(productDF['price'][i]),
        'brand': str(productDF['brand'][i]),
        'rating': int(productDF['Rating'][i]),
        'description': str(productDF['description'][i])
    }
    db.collection('products').document(str(productDF['product_id'][i])).set(data)
```

```python
docs = db.collection('Users').where('user_id', 'not-in', ['LfQFPMkqQ4bp07N0awGc4Gn6i2K2', 'bWhaebb57UUQ6Xz5bfTnTtf3UX92'] ).get()
for doc in docs:
    ratings = db.collection('Users').document(str(doc.id)).collection('ratings').get()
    for collection in ratings:
        item = collection.to_dict()
        product_ids = [item['productId']]
        rating = [item['rating']]
        user_id = [item['user_id']]
        tempUser_dataframe = pd.DataFrame({'user_id': user_id, 'product_id': product_ids, 'stars': rating})
        userRatings_dataframe = userRatings_dataframe.append(tempUser_dataframe, ignore_index=True)
userRatings_dataframe
```

```python
docs = db.collection('Users').where('user_id', 'in', [2323, 4862]).get()
for doc in docs:
    item = doc.to_dict()
    product_ids = item['product_id']
    rating = item['stars']
    user_id = [doc.id] * len(rating)
    combination_skin = [item['combination_skin']] * len(rating)
    dry_skin = [item['dry_skin']] * len(rating)
    normal_skin = [item['normal_skin']] * len(rating)
    oily_skin = [item['oily_skin']] * len(rating)
    sensitive_skin = [item['sensitive_skin']] * len(rating)
    skin_tone = [item['skin_tone']] * len(rating)
    skin_type = [item['skin_type']] * len(rating)
    eye_color = [item['eye_color']] * len(rating)
    hair_color = [item['hair_color']] * len(rating)
    tempUser_dataframe = pd.DataFrame({'user_id': user_id, 'product_id': product_ids, 'stars': rating})
    tempCharacter_dataframe = pd.DataFrame({'user_id': user_id, 'combination_skin': combination_skin, 'dry_skin' :dry_skin,
                                            'normal_skin': normal_skin, 'oily_skin': oily_skin, 'sensitive_skin': sensitive_skin,
                                            'skin_tone': skin_tone, 'skin_type': skin_type, 'eye_color': eye_color, 'hair_color': hair_color,
                                            'product_id': product_ids})
    user_dataframe = user_dataframe.append(temp_dataframe, ignore_index=True)
    character_dataframe = character_dataframe.append(tempCharacter_dataframe, ignore_index=True)
character_dataframe
```
